```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import pickle
```

```python
adds = pickle.load(open("diff_adds.pickle","rb"))
dels = pickle.load(open("diff_dels.pickle","rb"))
upds = pickle.load(open("diff_upds.pickle","rb"))
srcs = pickle.load(open("diff_srcs.pickle","rb"))
df_adds = pd.DataFrame.from_records(adds,index="date")
df_upds = pd.DataFrame.from_records(upds,index="date")
df_dels = pd.DataFrame.from_records(dels,index="date")
df_srcs = pd.DataFrame.from_records(srcs,index="date")
```

```python
df_adds.plot()
df_dels.plot()
df_upds.plot()
df_srcs["reagent"].plot()
print(df_srcs.info())
```
